### . Classify Movies Based on Ratings [Excellent, Good, and Average]

```python
data_df.columns
```

```python
def Rating(rating):
    if rating>=7 :
        return "Excellent"
    elif rating>=5 :
        return "good"
    else  :
        return "Average"
```

```python
data_df['rating_cat']=data_df['Rating'].apply(Rating)
```

### Find Unique Values From Genre 

```python
data_df.columns
```

```python
data_df['Genre']
```

```python
list1=[]
for value in data_df['Genre']:
    list1.append(value.split(','))
```

```python
list1
```

```python
one_d=[]
for item in list1 :
    for item1 in item :
        one_d.append(item1)
```

```python
one_d
```

```python
uni_list=[]
for item in one_d:
    if item not in uni_list:
        uni_list.append(item)
```

```python
uni_list
```
